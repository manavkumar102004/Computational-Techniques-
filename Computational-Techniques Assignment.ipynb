```python
# Program to find Mean, Median and Mode for grouped data

# Given data
age_groups = [(15,25), (25,35), (35,45), (45,55), (55,65)]
freq = [350, 375, 120, 80, 90]

# Step 1: Find mid-points (class marks)
mid = []
for a, b in age_groups:
    mid.append((a + b) / 2)

# Step 2: Mean
total_f = sum(freq)
total_fx = 0
for i in range(len(mid)):
    total_fx += mid[i] * freq[i]

mean = total_fx / total_f

# Step 3: Median
# Find cumulative frequency
cf = []
s = 0
for f in freq:
    s += f
    cf.append(s)

n = total_f
for i in range(len(cf)):
    if cf[i] >= n/2:
        median_class = i
        break

L = age_groups[median_class][0]
h = age_groups[median_class][1] - age_groups[median_class][0]
F = cf[median_class - 1] if median_class > 0 else 0
f = freq[median_class]

median = L + ((n/2 - F) / f) * h

# Step 4: Mode
modal_class = freq.index(max(freq))
L = age_groups[modal_class][0]
f1 = freq[modal_class]
f0 = freq[modal_class - 1] if modal_class > 0 else 0
f2 = freq[modal_class + 1] if modal_class < len(freq)-1 else 0
h = age_groups[modal_class][1] - age_groups[modal_class][0]

mode = L + ((f1 - f0) / (2*f1 - f0 - f2)) * h

# Print results
print("Mean  =", round(mean,2))
print("Median=", round(median,2))
print("Mode  =", round(mode,2))


```
